2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv.

In case if any product has less than 3 pages in search results then scrape all the products available under that product name. 

Details to be scraped are: 

"Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”.

In case, if any of the details are missing for any of the product then replace it by “-“.


```python
url=driver.find_elements_by_xpath("//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")


product_urls=[]


for r in range(0,3):
    for i in url:
        product_urls.append(i.get_attribute('href'))

        
try:        
    driver.find_element_by_xpath('//span[@class="s-pagination-item s-pagination-selected"]').click()
except NoSuchElementException:
    driver.find_element_by_xpath('//span//a[@class="s-pagination-item s-pagination-button"][2]').click()
```

```python
# Creating blank lists to store data after fetching
Brand_names = []
Product_names = []
Prices = []
Return_Policy = []
Delivery_Date = []
Availability = []
```

```python
from tqdm import tqdm
for link in tqdm(product_urls):
    driver.get(link)
           
    try:
        brand = driver.find_element_by_xpath("//*[@id='bylineInfo']").text
        Brand_names.append(brand)
    except NoSuchElementException:
        Brand_names.append('-')    
        
    try:
        name = driver.find_element_by_xpath("//*[@id='productTitle']").text
        Product_names.append(name)
    except NoSuchElementException:
        Product_names.append('-')
        
        
    try:
        price = driver.find_element_by_xpath("//*[@class='a-price a-text-price a-size-medium apexPriceToPay']").text
        Prices.append(price)
    except NoSuchElementException:
        Prices.append('-')
        
    try:
        exchange = driver.find_elements_by_xpath("//*[@id='icon-farm-container']/div/div[1]/span/div[2]/a")
        Return_Policy.append(exchange[0].text)        
    except NoSuchElementException:
        Return_Policy.append('-')
        
    try:
        delivery = driver.find_element_by_xpath("//div[@id='mir-layout-DELIVERY_BLOCK-slot-DELIVERY_MESSAGE']/b").text
        Delivery_Date.append(delivery)
    except NoSuchElementException:
        Delivery_Date.append('-')
        
    try:
        avail = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price'] | //span[@class='a-size-medium a-color-success'] ").text
        Availability.append(avail)
    except NoSuchElementException:
        Availability.append('-')
        


        
time.sleep(3)        
```

```python
# creating empty lists to collect the required information
hostel_name = []
distance = []
pvt_prices = []
dorms_price = []
rating = []
reviews = []
over_all = []
facilities = []
description =[]
product_url = []
```

```python


page=driver.find_elements_by_xpath("//div[@class = 'pagination-item pagination-current']")
for i in page:
    i.click()
    time.sleep(4)
    
    # fetching hostel name
    try:
        name = driver.find_elements_by_xpath("//h2[@class='title title-6']")
        for i in name:
            hostel_name.append(i.text)
    except NoSuchElementException:
        hostel_name.append('-')
        
    # fetching the distance from city centre    
    try:
        dist = driver.find_elements_by_xpath("//div[@class='subtitle body-3']//a//span[1]")
        for i in dist:
            distance.append(i.text.replace('Hostel - ',''))
    except NoSuchElementException:
        distance.append('-')
        
    for i in driver.find_elements_by_xpath("//div[@class='prices-col']"):
        
    # fetching the privates from price details
        try:
            pvt_price = driver.find_element_by_xpath("//a[@class='prices']//div[1]//div")
            pvt_prices.append(pvt_price.text)
        except NoSuchElementException:
            pvt_prices.append('-')
            
    # fetching the dorms from price information
    for i in driver.find_elements_by_xpath("//div[@class='prices-col']"):
        try:
            dorms = driver.find_element_by_xpath("//a[@class='prices']//div[2]//div")
            dorms_price.append(dorms.text)
        except NoSuchElementException:
            dorms_price.append('-')
            
    # fetching all the facilities
    try:
        fac1 = driver.find_elements_by_xpath("//div[@class='has-wifi']")
        fac2 = driver.find_elements_by_xpath("//div[@class='has-sanitation']")
        for i in fac1:
            for j in fac2:
                facilities.append(i.text +', '+ j.text )
    except NoSuchElementException:
        facilities.append('-')
        
    # fetching the urls of each hostel
    p_url = driver.find_elements_by_xpath("//div[@class='prices-col']//a[2]")
    for i in p_url:
        product_url.append(i.get_attribute('href'))

for i in product_url:
    driver.get(i)
    time.sleep(3)
    
     
    
    # fetching the ratings for the hostels
    try:
        rat = driver.find_element_by_xpath("//div[@class='score orange big' or @class='score gray big']")
        rating.append(rat.text)
    except NoSuchElementException:
        rating.append('-')
        
    # fetching the total reviews        
    try:
        rws = driver.find_element_by_xpath("//div[@class='reviews']")
        reviews.append(rws.text.replace('Total Reviews',''))
    except NoSuchElementException:
        reviews.append('-')
        
    # fetching the overall review
    try:
        overall_rw = driver.find_element_by_xpath("//div[@class='keyword']//span")
        over_all.append(overall_rw.text)
    except NoSuchElementException:
        over_all.append('-')
        
    # fetching the property description 
    try:
        disc = driver.find_element_by_xpath("//div[@class='content']")
        description.append(disc.text)
    except NoSuchElementException:
        over_all.append('-')



```

```python
# Creating a DATA FRAME to store fetched data
Hostel = pd.DataFrame({})
Hostel['Hostel name'] = hostel_name
Hostel['Distance from city centre'] = distance
Hostel['Rating'] = rating
Hostel['Total reviews'] = reviews
Hostel['Overall review'] = over_all
Hostel['Privates from price'] = pvt_prices
Hostel['Dorms from price'] = dorms_price
Hostel['Facilities'] = facilities 
Hostel['Property Description'] = description
Hostel
```
